**A Systemic Review on Stopword Removal Algorithms**

Los stopwords son palabras que se repiten mucho y que contienen poca información que por lo general no es requerida. El eliminar los stopwords, además de reducir el tamaño de los textos, aumenta la velocidad de ejecución y la precisión. Los stopwords sólo tienen una función sintáctica y no tienen poder predictivo ya que no indican el tema de un texto. Existen varios métodos para remover stopwords de un texto:
* Método clásico: remover palabras que se enceuntren en una lista predeterminada de stopwords.
* Métodos basados en la Ley de Zipf: Remover las palabras más frecuentes, menos frecuentes y palabras con una baja "inverse document frequency (IDF).
* Método de la información mutua: método supervisado en el que se calcula la información mutua entre una palabra y el tipo de documento (positivo, negativo), indicando de esta manera qué tanta información provée la palabra sobre el texto. Una baja información mutua sugiere que la palabra tienen un bajo poder de discriminación por lo que se debe de eliminar.
* Term based Random Sampling (TBRS): Se eliminan aquellas palabras que sean las menos informativas al eliminar las palabras duplicads. Se utiliza una fórmula para calcular la frecuencia de una palabra en un texto.

Otros autores han sugerido una nueva lista de stopwords basada en la entropía de las palabras. Se introdujo la optimización de una lista de stopwords basada en búsqueda estocástica combinando k-means clustering. Diferentes autores han combinado varios de los métodos para eliminar stopwords, y el procesamiento y precisión de las predicciones han aumentado así como que el tiempo de procesamiento ha disminuido considerablemente. 

**Pulling Out the Stops: Rethinking Stopword Removal for Topic Models**

El eliminar ciertas palabras asociadas con  el tema principal del texto no trae muchos beneficios. Es mejor eliminar solamente aquellas palabras más frecuentes. Se conluye que eliminar los stopwords sí trae ciertos beneficios sin embargo la remoción de palabras no parece mejorar consistentemente la habilidad del modelo para aprender , más bien reduce el tiempo en el que se hacen las inferencias. Se concluye que los stopwords no afectan significativamente el proceso de inferencia para otros términos. 

**Normalizing Text using Language Modelling based on Phonetics and String Similarity**

Las personas utilizamos un lenguaje infromal al comunicarnos por medio de las redes sociales y el internet. Para poder normalizar este lenguaje, se utiliza el modelo BERT para predecir las mascaras que corresponden a las palabras que no estan normalizadas.

Para identificar las palabras que requieren normalizacion se siguen estos pasos:

  1. Primero se identifican las palabras que ya se encuentran en el diccionario NLTK
  2. Encontrar los acronimos, por ejemplo "GM" por "Good Morning", "ROFL" por "Rolling On the Floor Laughing"
  3. Encontrar las contracciones, por ejemplo "hasn't" por "Has not", "you've " por "You have"
  4. Encontrar los nombres propios, estos no deben ser modificados. Para poder identificarlos se usa NER (Named Entity Recognition) con una red neuronal residual bidireccional LSTM entrenada en ELMo embeddings.
  5. Identificar palabras no normalizadas, las cuales no pertenecen a ninguna de las categorias anteriores.

Para convertir las palabras no normalizadas se utilizan dos diferentes metodos:

**1. OOV (Out-of-Vocabulary) Masking**

Este metodo consiste en utilizar la prediccion BERT unicamente en aquellas palabras que no se encuentran en el vocabulario.

1. Se cambian las palabras no normalizadas en el enunciado por un token que funciona como placeholder.

2. Se asigna un puntaje a las posibles palabras que podrian completar el enunciado, este puntaje esta determinado por 3 componentes:

 a. Probabilidad basada en el contexto.

 b. Similaridad en la cadena.

 c. Similaridad en la fonetica.

3. Reemplazar el token por la palabra con mejor puntaje. Si el mejor puntaje es menor a 0.25, la palabra se deja sin normalizar.

**2. Word-by-Word Masking**

Uno de los problemas del metodo anterior, es que asume que todas las palabras no normalizadas son palabras OOV, sin embargo a veces puede ser el caso de encontrar palabras que se encuentran dentro del diccionario pero que necesitan ser normalizadas, por ejemplo: "wit" en lugar de "with" o "cud" en lugar de "could".

**3. Realiza un proceso de tokenización del conjunto de entrenamiento y define tu vocabulario usando
el conjunto de entrenamiento. Deberás indicar la frecuencia mínima requerida de cada token para
que sea parte del vocabulario. Además usar la opción OOV_token en el proceso de tokenización

entrenamiento en secuencia de enteros, aplica el resultado obtenido al conjunto de validación.**

```python
vocabulario = Counter()
for question in x_train:
  vocabulario.update(question)
```

```python
print(len(vocabulario))
print(vocabulario.most_common(92000))
```

```python
# Gráfica de la frecuencia de las palabras
plt.plot(list(np.arange(len(vocabulario))), list(vocabulario.values()), color='blue')
plt.show()
```

```python
# Se eliminarán aquellas palabras cuya frecuencia en el conjunto de entrenamiento es muy pequeña
min_freq = 50
reduced_dict = [{k:c} for k,c in vocabulario.items() if c >= min_freq]

new_dict = {}
for word_tuple in reduced_dict:
  new_dict.update(word_tuple)

total_num_words = len(new_dict)
print(len(new_dict))
print("% Reducción del diccionario", (1-len(new_dict)/len(vocabulario))*100)
```

```python
plt.plot(list(np.arange(len(new_dict))), list(new_dict.values()), color='blue')
plt.show()
```

```python
def remove_words(x_set,new_vocabulary_dict):
  """
  x_set: a list of lists
  new_vocabulary_dict: dictionary with the new words as keys and the number of words as value
  """
  new_set = []
  for sentence in x_set:
    new_set.append([word for word in sentence if word in new_vocabulary_dict and word != ''])
  return new_set

```

```python
x_train_reduced_arr = remove_words(x_train,new_dict)
x_val_reduced_arr = remove_words(x_val,new_dict)
x_test_reduced_arr = remove_words(x_test,new_dict)
```

```python
print(len(x_train))
print(len(x_train_reduced_arr))
print(len(x_val))
print(len(x_val_reduced_arr))
```

```python
def generate_sentences_from_arrays(x_set):
  new_set = []
  for sentence in x_set:
    new_set.append([" ".join(sentence)])
  return new_set
```

```python
x_train_reduced = generate_sentences_from_arrays(x_train_reduced_arr)
x_val_reduced = generate_sentences_from_arrays(x_val_reduced_arr)
```

Tokenización

```python
tokenizer = Tokenizer(num_words=total_num_words,oov_token='UNK')
tokenizer.fit_on_texts(x_train_reduced_arr)
x_train_sequences = tokenizer.texts_to_sequences(x_train_reduced_arr)
x_val_sequences = tokenizer.texts_to_sequences(x_val_reduced_arr)
x_test_sequences = tokenizer.texts_to_sequences(x_test_reduced_arr)
```

```python
n_epochs = 10
vocab_size = tokenizer.num_words
sent_length = pad_length
```

```python
model = Sequential()
model.add(Embedding(input_dim=vocab_size,
                    output_dim=128,
                    input_length=sent_length))

model.add(Bidirectional(SimpleRNN(units=128)))

model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
```

```python
model2 = Sequential()
model2.add(Embedding(input_dim=vocab_size,
                    output_dim=128,
                    input_length=sent_length))

model2.add(Bidirectional(SimpleRNN(units=512)))

model2.add(Flatten())
model2.add(Dense(1, activation='sigmoid'))
```

```python
model3 = Sequential()
model3.add(Embedding(input_dim=vocab_size,
                    output_dim=64,
                    input_length=sent_length))

model3.add(Bidirectional(SimpleRNN(units=64, return_sequences=True)))
model3.add(Bidirectional(SimpleRNN(units=64, return_sequences=True)))
model3.add(Bidirectional(SimpleRNN(units=64)))

model3.add(Flatten())
model3.add(Dense(1, activation='sigmoid'))
```

```python
def plot_results(H, n_epochs):
  plt.style.use("ggplot")
  plt.figure()
  plt.plot(np.arange(0,n_epochs),H.history["accuracy"],label="train_acc")
  plt.plot(np.arange(0,n_epochs),H.history["val_accuracy"],label="val_acc")
  plt.title("Training and Validation Accuracy")
  plt.xlabel("Epoch")
  plt.ylabel("Accuracy")
  plt.legend()

  plt.figure()
  plt.plot(np.arange(0,n_epochs),H.history["loss"],label="train_loss") 
  plt.plot(np.arange(0,n_epochs),H.history["val_loss"],label="val_loss")
  plt.title("Training and Validation Loss")
  plt.xlabel("Epoch")
  plt.ylabel("Loss")
  plt.legend()

  plt.show()
```

**8. Muestra la matriz de confusión del modelo, utilizando los datos del conjunto de validación.**

```python
def conf_mat(model):
  y_pred = model.predict(x_val_pad)
  y_pred = tf.greater(y_pred,0.5)
  # cm = confusion_matrix(y_val, y_pred.argmax(axis=1))
  cm = confusion_matrix(y_val, y_pred)
  labels = ["Android", "iOS"]
  pd_cm = pd.DataFrame(cm, index = labels, columns=labels)
  plt.figure(figsize=(10,5))
  g = sn.heatmap(pd_cm, annot=True, cmap="Blues", fmt="d", annot_kws={"fontsize":15})
  g.set_xticklabels(g.get_xmajorticklabels(), fontsize = 13)
  g.set_yticklabels(g.get_ymajorticklabels(), fontsize = 13)
  plt.yticks(rotation=0)
  plt.xlabel('Predicted labels')
  plt.ylabel('True labels')
  plt.show()
```
