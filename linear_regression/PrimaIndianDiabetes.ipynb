```r
library(class)
library(gmodels)
```

## Read Dataset

```r
df<-read.csv("0500_pima_indians_diabetes_data.csv", header=FALSE)
```

## Visualize Dataset

```r
normalize <- function(x) {
    return ((x - min(x)) / (max(x) - min(x)))
}
df_n <- as.data.frame(lapply(df[1:8], normalize))
summary(df_n)
```

```r
boxplot(df_n, las=2)
```

## Train model with 60% train data and 40% test data and k = 10

```r
df_train <- df_n[1:460,]
df_test <- df_n[461:768,]
df_train_labels <- df[1:460, 9]
df_test_labels <- df[461:768, 9]
```

```r
df_test_pred <- knn(train = df_train, test = df_test, cl = df_train_labels, k=10)
```

### Confussion matrix shows 76% accuaracy

```r
CrossTable(x = df_test_labels, y = df_test_pred, prop.chisq = FALSE)
```
