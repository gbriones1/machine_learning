```python
import math

import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    "x1": [0.15, -0.1, 0, -0.25, 0.05],
    "x2": [3500, 3400, 4000, 3900, 3200],
    "y": [8, 5, 3, 1, 7]
})
print(df)
```

```python
def minmax_normalization(array, a=0, b=1):
    result = []
    _min = min(array)
    _max = max(array)
    _range = _max - _min
    for element in array:
        result.append(a + ((element - _min) * (b - a))/_range)
    return result

def mean_normalization(array):
    result = []
    _range = max(array) - min(array)
    _average = np.average(array)
    for element in array:
        result.append((element - _average)/_range)
    return result

def standarization(array):
    result = []
    std = np.std(array)
    _average = np.average(array)
    for element in array:
        result.append((element - _average)/std)
    return result
    

print("X1 scaled: ", mean_normalization(df["x1"]))
print("X2 scaled: ", standarization(df["x2"]))
print("Y scaled: ", minmax_normalization(df["y"], -10, 20))
```

2. Existen varias formas de medir la distancia entre puntos. MÃ¡s adelante requeriremos algunas de ellas para algunos algoritmos de clasificaciÃ³n. Por el momento investiga cÃ³mo se definen las distancias o mÃ©tricas indicadas en cada inciso y utilÃ­zalas para encontrar la distancia entre los puntos ğ´=(2,1) y ğµ=(3,4) de la figura dada.

a. MÃ©trica Euclidiana.

b. MÃ©trica Manhattan.

c. MÃ©trica Chebyshev

d. MÃ©trica Minkowski de orden ğ‘ = 0.7

```python
def euclidean(a, b):
    return math.sqrt((b[0] - a[0])**2 + (b[1] - a[1])**2)

def manhattan(a, b):
    return abs(b[0] - a[0]) + abs(b[1] - a[1])

def chevysheb(a, b):
    return max((abs(b[0] - a[0]), abs(b[1] - a[1])))

def minkowski(a, b, p):
    return abs(b[0] - a[0])**p + abs(b[1] - a[1])**p
    

a = (2, 1)
b = (3, 4)
p = 0.7
print("A:", a)
print("B:", b)
print("Euclidean distance: ", euclidean(a, b))
print("Manhattan distance: ", manhattan(a, b))
print("Chevysheb distance: ", chevysheb(a, b))
print("Minkowski distance with p=%f: " % p, minkowski(a, b, p))
```

```python
Q = np.array([[1, -3, 1], [-3, 9, -3], [1, -3, 1]])
print("Q:\n", Q, "\n")

q = np.array([[-1, 3, -1]])
print("q:\n", q, "\n")

qt = q.transpose()
print("qT:\n", qt, "\n")

qtq = np.dot(qt, q)
print("qTq:\n", qtq, "\n")

print("Comparison between Q and qtq:\n", qtq == Q)
```

```python
def cross_correlation(image, kernel, convolution=False, padding="zeroes"):
    image_row, image_col = image.shape
    kernel_row, kernel_col = kernel.shape

    output = np.zeros(image.shape)

    pad_height = int((kernel_row - 1) / 2)
    pad_width = int((kernel_col - 1) / 2)
    
    padded_image = image
    
    if padding:
        padded_image = np.zeros((image_row + (2 * pad_height), image_col + (2 * pad_width)))
        padded_image[pad_height:padded_image.shape[0] - pad_height, pad_width:padded_image.shape[1] - pad_width] = image
        if padding == "replica":
            if pad_height:
                padded_image[0:pad_height, :] = padded_image[pad_height:pad_height+1, :]
                padded_image[-pad_height:, :] = padded_image[-pad_height-1:-pad_height, :]
            if pad_width:
                padded_image[:, 0:pad_width] = padded_image[:, pad_width:pad_width+1]
                padded_image[:, -pad_width:] = padded_image[:, -pad_width-1:-pad_width]
        elif padding == "symmetry":
            if pad_height:
                padded_image[0:pad_height, :] = np.flip(padded_image[pad_height:pad_height*2, :], axis=0)
                padded_image[-pad_height:, :] = np.flip(padded_image[-pad_height*2:-pad_height, :], axis=0)
            if pad_width:
                padded_image[:, 0:pad_width] = np.flip(padded_image[:, pad_width:pad_width*2], axis=1)
                padded_image[:, -pad_width:] = np.flip(padded_image[:, -pad_width*2:-pad_width], axis=1)
        elif padding == "cyclic":
            if pad_height:
                padded_image[0:pad_height, :] = padded_image[-pad_height*2:-pad_height, :]
                padded_image[-pad_height:, :] = padded_image[pad_height:pad_height*2, :]
            if pad_width:
                padded_image[:, 0:pad_width] = padded_image[:, -pad_width*2:-pad_width]
                padded_image[:, -pad_width:] = padded_image[:, pad_width:pad_width*2]

    if convolution:
        kernel = np.rot90(kernel, 2)

    for row in range(image_row):
        for col in range(image_col):
            output[row, col] = np.sum(kernel * padded_image[row:row + kernel_row, col:col + kernel_col])
    return output


H = np.array([
    [10, 20, 0, 30],
    [20, 0, 10, 20],
    [10, 0, 40, 0],
    [15, 40, 0, 30]
])
print("ğ»:\n", H, "\n")

HxQ = cross_correlation(H, Q, convolution=True, padding="symmetry")
print("ğ» âˆ— ğ‘„:\n", HxQ, "\n")

Hxq = cross_correlation(H, q, convolution=True, padding="symmetry")
print("ğ» âˆ— ğ‘:\n", Hxq, "\n")

Hxqxqt = cross_correlation(Hxq, qt, convolution=True, padding="symmetry")
print("(ğ» âˆ— ğ‘) âˆ— ğ‘T:\n", Hxqxqt, "\n")
```

4. Mediante las fÃ³rmulas correspondientes vistas en clase y a partir de la imagen ğ¼ mostrada mÃ¡s abajo (se muestran solamente las columnas 15, 16, 17 y los renglones 6, 7 ,8), interpolar el valor ğ¼(15.85, 7.40), usando:

a. InterpolaciÃ³n mediante el vecino mÃ¡s cercano.

b. InterpolaciÃ³n bilineal.

| | | |
|-----|-----|----|
| 110 | 125 | 98 |
| 106 | 120 | 95 |
| 100 | 110 | 90 |

```python
def nn_interpolation(image, x0, y0):
    return image[round(x0)][round(y0)]

def bl_interpolation(image, x0, y0):
    u, a = math.modf(x0)
    v, b = math.modf(y0)
    a = int(a)
    b = int(b)
    i = 0
    i += image[a][b] * (1 - u) * (1 - v)
    i += image[a+1][b] * u * (1 - v)
    i += image[a][b+1] * (1 - u) * v
    i += image[a+1][b+1] * u * v
    return i

I = np.array([
    [110, 125, 98],
    [106, 120, 95],
    [100, 110, 90]
])
print("Image[15:18, 6:9]:\n", I, "\n")

print("Nearest Neighbor Interpolation of (15.85, 7.40):", nn_interpolation(I, 15.85-15, 7.40-6))
print("Bilinear Interpolation of (15.85, 7.40):", bl_interpolation(I, 15.85-15, 7.40-6))
```

6. Aplicar una ecualizaciÃ³n uniforme a la imagen de 3-bits en niveles de grises
dada. Utilizar la funciÃ³n acumulada uniforme y la funciÃ³n piso para
redondear.

| | | | |
|---|---|---|---|
| 6 | 6 | 7 | 5 |
| 6 | 7 | 5 | 4 |
| 6 | 5 | 3 | 4 |
| 7 | 5 | 4 | 3 |

```python
def equalize(matrix, bits):
    result = np.zeros(matrix.shape)
    new_values = {}
    nm = matrix.shape[0] * matrix.shape[1]
    acc = 0
    for g in range(bits):
        ng = np.count_nonzero(matrix == g)
        acc += ng/nm
        new_values[g] = math.floor(acc * bits)
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            result[i][j] = new_values[matrix[i][j]]
    return result
            

image = np.array([
    [6, 6, 7, 5],
    [6, 7, 5, 4],
    [6, 5, 3, 4],
    [7, 5, 4, 3]
])

equilized = equalize(image, 8)

print("Image:\n", image, "\n")
print("Equilized image:\n", equilized, "\n")
```
